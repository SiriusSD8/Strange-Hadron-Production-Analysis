```python
import matplotlib.pyplot as plt
import numpy as np

Pt = np.loadtxt("./(6)7.7GeV-AL_KoS(0-5%).txt", usecols = 0)
AL_KoS = np.loadtxt("./(6)7.7GeV-AL_KoS(0-5%).txt", usecols = 1)
xerr = np.loadtxt("./(6)7.7GeV-AL_KoS(0-5%).txt", usecols = 2)
yerr = np.loadtxt("./(6)7.7GeV-AL_KoS(0-5%).txt", usecols = 2)

deg = 2
parameters2 = np.polyfit(Pt, AL_Kos, deg)
model = parameters2[0]*Pt**2 + parameters2[1]*Pt + parameters2[2]

plt.errorbar(Pt, AL_KoS, xerr, yerr, fmt = 'o', color = 'blue', ecolor = 'green', linewidth = 1)
plt.plot(Pt, model, '-', color = 'black', alpha = 0.4)
plt.xlabel("Pt [GeV/c]")
plt.ylabel("Λ-/KoS")
plt.title("Pt [GeV/c] Vs. Λ-/KoS at 7.7GeV")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

AL_KoS7_7 = np.loadtxt("./(6)7.7GeV-AL_KoS(0-5%).txt", usecols = 1)
AL_KoS11_5 = np.loadtxt("./(7)11.5GeV-AL_KoS(0-5%).txt", usecols = 1)
AL_KoS19_6 = np.loadtxt("./(8)19.6GeV-AL_KoS(0-5%).txt", usecols = 1)
AL_KoS27 = np.loadtxt("./(9)27GeV-AL_KoS(0-5%).txt", usecols = 1)
AL_KoS39 = np.loadtxt("./(10)39GeV-AL_KoS(0-5%).txt", usecols = 1)

AL_KoSF_stat, AL_KoSp_val = stats.f_oneway(AL_KoS7_7, AL_KoS19_6, AL_KoS27)
print("P-value when comparing AntiLambda/KoS at GeV = 7.7, 19.6, 27: ",AL_KoSp_val)

AL_KoSF_stat, AL_KoSp_val = stats.f_oneway(AL_KoS7_7, AL_KoS19_6, AL_KoS27, AL_KoS39)
print("P-value when comparing AntiLambda/KoS at GeV = 7.7, 19.6, 27, 39: ",AL_KoSp_val)

AL_KoSF_stat, AL_KoSp_val = stats.f_oneway(AL_KoS11_5, AL_KoS19_6, AL_KoS27)
print("P-value when comparing AntiLambda/KoS at GeV = 11.5, 19.6, 27: ",AL_KoSp_val)

AL_KoSF_stat, AL_KoSp_val = stats.f_oneway(AL_KoS11_5, AL_KoS19_6, AL_KoS27, AL_KoS39)
print("P-value when comparing AntiLambda/KoS at GeV = 11.5, 19.6, 27, 39: ",AL_KoSp_val)
```
