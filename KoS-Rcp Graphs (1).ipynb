```python
import matplotlib.pyplot as plt
import numpy as np

KoS = np.loadtxt("./7.7GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 0)
RcP = np.loadtxt("./7.7GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)
xerr = np.loadtxt("./7.7GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 2)
yerr = np.loadtxt("./7.7GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 4)

deg = 2
parameters2 = np.polyfit(KoS, RcP, deg)
model = parameters2[0]*KoS**2 + parameters2[1]*KoS + parameters2[2]

plt.errorbar(KoS, RcP, xerr, yerr, fmt = 'o', color = 'blue', ecolor = 'green', linewidth = 1)
plt.plot(KoS, model, '-', color = 'black', alpha = 0.4)
plt.xlabel("Pt [GeV/c]")
plt.ylabel("KoS(RcP)")
plt.title("Pt [GeV/c] Vs. KoS(RcP) at 7.7GeV")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

RcP7_7 = np.loadtxt("./7.7GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)
RcP11_5 = np.loadtxt("./11.5GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)
RcP19_6 = np.loadtxt("./19.6GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)
RcP27 = np.loadtxt("./27GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)
RcP39 = np.loadtxt("./39GeV-KoS-Vs-RcP(0-5%_40-60%).txt", usecols = 1)

alpha = 0.01
CV = stats.norm.ppf(1 - alpha)
print("Critical Value:", CV)

RcP_tstat, RcP_pval = stats.ttest_ind(RcP7_7, RcP19_6)

if RcP_tstat > CV:
    print("(7.7-19.6GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)

RcP_tstat, RcP_pval = stats.ttest_ind(RcP7_7, RcP27)

if RcP_tstat > CV:
    print("(7.7-27GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)
    
RcP_tstat, RcP_pval = stats.ttest_ind(RcP7_7, RcP39)

if RcP_tstat > CV:
    print("(7.7-39GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)
    
RcP_tstat, RcP_pval = stats.ttest_ind(RcP11_5, RcP19_6)

if RcP_tstat > CV:
    print("(11.5-19.6GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)

RcP_tstat, RcP_pval = stats.ttest_ind(RcP11_5, RcP27)

if RcP_tstat > CV:
    print("(11.5-27GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)

RcP_tstat, RcP_pval = stats.ttest_ind(RcP11_5, RcP39)

if RcP_tstat > CV:
    print("(11.5-39GeV)Null Hypothesis is rejected, there is a difference. Tstat:", RcP_tstat)
```
